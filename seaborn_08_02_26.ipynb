#import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Import Data

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "ComputerSales.csv"

cs_df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "priyanshgoantiya/computersales",
  file_path,
)
```

```python
cs_df.head()
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

#Distribution Plots

```python
sns.displot(crash_df['not_distracted'], kde=False, bins=25)
```

#Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

#KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

#Pair Plots

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
tips_df.shape
```

```python
tips_df
```

#Rug Plots

```python
sns.rugplot(tips_df['tip'])
```

#Styling

```python
sns.set_style('white')
plt.figure(figsize=(8,4))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=False, bottom=False)
```

#Bar Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips_df, estimator=np.median, hue='sex')
```

#Count Plot

```python
sns.countplot(x='sex',data=tips_df, hue='sex')
```
